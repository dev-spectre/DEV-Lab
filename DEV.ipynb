```python
import numpy as np
```

```python
# 1D Array
arr_1d = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr_1d)
```

```python
# 2D Array
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", arr_2d)
```

```python
# Array with specific shape and values
zeros = np.zeros((2, 3))
ones = np.ones((3, 2))
identity = np.eye(3)

print("Zeros Array:\n", zeros)
print("Ones Array:\n", ones)
print("Identity Matrix:\n", identity)
```

```python
# Element-wise operations
arr_add = arr_1d + 10
arr_mult = arr_2d * 2

print("Array after addition:\n", arr_add)
print("Array after multiplication:\n", arr_mult)
```

```python
# Matrix operations
matrix_mult = np.dot(arr_2d, np.array([[1, 0], [0, 1], [1, 0]]))
print("Matrix Multiplication:\n", matrix_mult)
```

```python
import pandas as pd
```

```python
# Creating DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
print("DataFrame:\n", df)
```

```python
# Basic statistics
print("Statistics:\n", df.describe(include='all'))
```

```python
# Accessing specific columns
print("Names:\n", df['Name'])
```

```python
# Filtering rows
filtered_df = df[df['Age'] > 28]
print("Filtered DataFrame:\n", filtered_df)
```

```python
# Adding a new column
df['Country'] = 'USA'
print("DataFrame with New Column:\n", df)
```

```python
import matplotlib.pyplot as plt
```

```python

# Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='blue', linestyle=':')
plt.title('Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.show()
```

```python

# Histogram
data = np.random.randn(1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python

# Bar Plot
categories = ['A', 'B', 'C']
values = [10, 20, 15]

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='green')
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python

import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Create a map
m = Basemap(projection='merc', llcrnrlat=32, urcrnrlat=42, llcrnrlon=-125, urcrnrlon=-114, resolution='i')

# Draw map features
m.drawcoastlines()
m.drawcountries()
m.drawstates()
m.fillcontinents(color='lightgreen', lake_color='aqua')
m.drawmapboundary(fill_color='aqua')

cities = {'Los Angeles': (34.05, -118.25),
          'San Francisco': (37.77, -122.42),
          'San Diego': (32.72, -117.16),
          'Sacramento': (38.58, -121.49)}

# Plot cities
for city, (lat, lon) in cities.items():
    x, y = m(lon, lat)
    m.plot(x, y, 'bo', markersize=8)
    plt.text(x, y, city, fontsize=12, ha='right')

plt.title('Major Cities in California')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np

# Create a map
m = Basemap(projection='lcc', lat_0=35, lon_0=-100, width=1E6, height=1.2E6, resolution='i')

# Draw map features
m.drawcoastlines()
m.drawcountries()
m.drawstates()
# Generate synthetic temperature data
lons = np.linspace(-110, -90, 100)
lats = np.linspace(30, 45, 100)
lons, lats = np.meshgrid(lons, lats)
temps = np.sin(lats / 5.0) * np.cos(lons / 5.0) * 20 + 10  # Synthetic temperature data

# Convert to map projection coordinates
x, y = m(lons, lats)

# Plot the temperature data
cs = m.contourf(x, y, temps, cmap='coolwarm')
m.colorbar(cs, label='Surface Temperature (°C)')

plt.title('Surface Temperature')
plt.show()
```
